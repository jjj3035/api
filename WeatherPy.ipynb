```python
# Dependencies
import matplotlib.pyplot as plt
import requests
import pandas as pd
from config import api_key
from citipy import citipy
import random
import seaborn as sns
import csv
```

```python
#Inititalize variables
random.seed()
counter = 0
cities = []
```

```python
# create a list of cities based on cordinates
while counter < 600:
    lat, lon = random.randint(-90, 90), random.randint(-180, 180)
    city = citipy.nearest_city(lat, lon)
    if city not in cities:
        #print city
        cities.append(city)
        counter = counter + 1
    else:
        continue            
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"
```

```python
# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
#create list to store city, country, lat, long, Date, temp, humidity, cloudiness, wind speed
city_name=[]
country=[]
lat=[]
lon=[]
date=[]
temp=[]
humidity=[]
cloud=[]
wind_speed=[]

for city in cities:
    try: 
        current_weather = requests.get(query_url + city.city_name + ',' + city.country_code).json()
        print (query_url + city.city_name + ',' + city.country_code)
        print ('city name:', current_weather['name'], ',' , 'city id:', current_weather['id'])
        city_name.append(current_weather['name'])
        country.append(current_weather['sys']['country'])
        lat.append(current_weather['coord']['lat'])
        lon.append(current_weather['coord']['lon'])
        date.append(current_weather['dt'])
        temp.append(current_weather['main']['temp_max'])
        humidity.append(current_weather['main']['humidity'])
        cloud.append(current_weather['clouds']['all'])
        wind_speed.append(current_weather['wind']['speed'])
        
    except Exception as err:
        print (err)
        pass
```

```python
#Create Python Data Frame for the City weather data objects
city_data = pd.DataFrame({'City':city_name, 'Cloudiness':cloud, 'Country':country, 'Date':date, 'Humidity':humidity,
                         'Lat':lat, 'Lng':lon, 'Max Temp':temp, 'Wind Speed': wind_speed})
city_data.head()
```

```python
#Export City Weather Data to CSV File
city_data.to_csv('city_data.csv', encoding='utf-8', index=False)
```

```python
#Scatter Plot Latitude vs Temperature (F). 

sns.regplot(x=city_data['Lat'], y=(city_data['Max Temp']*1.8)+32, fit_reg=False)

# Incorporate the other graph properties
plt.title("City Latitude vs Max Temperature (03/22/2018)")
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)
plt.savefig('lat_vs_temp')
plt.show()

```
